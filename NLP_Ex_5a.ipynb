```python
import nltk
from nltk.util import ngrams
import pandas as pd
```

```python
with open('/content/drive/MyDrive/NLP DATASET/t8.shakespeare.txt', 'r') as f:
    data = f.read()

# #Convert the data to a dataframe
# df = pd.DataFrame({'text': [data]})
# df.head()
```

```python
words = data.split()
```

```python
trigrams = ngrams(words, 3)
```

```python
for trigram in trigrams:
    print(trigram)
```
